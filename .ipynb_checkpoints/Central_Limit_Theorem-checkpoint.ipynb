```python
import numpy as np
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
pop_data=[np.random.randint(10,1000) for i in range(10000)]
pop_df=pd.DataFrame({"pop_data":pop_data})
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="pop_data",data=pop_df)
plt.show()
```

```python
sample_mean=[]
for no_sample in range(50):
    sample_data=[]
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))    
```

```python
Sample_m=pd.DataFrame({"sample_mean":sample_mean})
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="sample_mean",data=Sample_m)
plt.show()
```
